```python
# Importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/train.csv')
```

# Groupby methods

```python

grp=df.groupby('Embarked')   # We have group on basis of embarked

#------------------------------------------------------------

# To count number of group
print('Total number of groups:-',end='')
print(grp.ngroups)

```

```python
# To view first row of each group
grp.first()
```

```python
# To get info of specific group
grp.get_group('C')
```

```python
#To get aggregate info about another column

grp1=df.groupby('Embarked')['Age'].mean() 
print(grp1)

grp1=df.groupby('Embarked')['Age'].max()
print(grp1)
```

```python
# Multiple aggregate function

df.groupby('Embarked')['Age'].agg([np.mean,np.median,np.std])
```

**Transform in Groupby**

In groupby we can easily replace each and every column on basis of its group.

```python
grp3=df.groupby('Embarked')['Age'].mean()
print('Mean of each group:-')
print(grp3)
```

* Now creating a columns where each value will be replaced with group value i-e if in a row embarked is 1 the value of that columns would be 30

```python
df['mean_value_of_each_group']=df.groupby('Embarked')['Age'].transform(np.mean)
df['mean_value_of_each_group'].value_counts()
```

**Filter in groupby**
* We can filter inside a groupby on basis of some condition


**It is used where we have to filter on basis of group by method**

```python
df.groupby('Embarked')['Age'].mean()


# Now suppose I want to view only row with value greater than its mean value of group

df.groupby('Embarked').filter(lambda x: x['Age'].mean()>30)
```

```python
# Now suppose I want to view only row with value greater than its mean value of group

df.groupby('Embarked').filter(lambda x: x['Age'].mean()>30)
```

**Apply ranking to your value**

```python
#  Suppose you want to rank on basis of Age

df['Age_rank']=df['Age'].rank(ascending=False)
df.loc[df['Age']==80,'Age_rank']  # So Age of 80 is rank 1
```
